# 2. Check for Missing or Null Values

```python
# Check amazon.csv for missing data

import pandas as pd

def check_missing_data(file_path):
    try:
        df = pd.read_csv(file_path)
        missing_data = df.isnull().sum()
        return missing_data
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

# Example usage (replace 'amazon.csv' with the actual file path)
missing_info = check_missing_data('amazon.csv')

if missing_info is not None:
  missing_info

missing_info
```

# 2.1 Find rows with missing 'rating_count' and return rows.

```python
# Find and return rows with missing data in column 'rating_count'

def get_rows_with_missing_rating_count(file_path):
    try:
        df = pd.read_csv(file_path)
        # Find rows where 'rating_count' is missing
        missing_rating_rows = df[df['rating_count'].isnull()]
        return missing_rating_rows
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
        return None
    except KeyError:
        print(f"Error: 'rating_count' column not found in the file.")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

# Example usage
missing_rating_data = get_rows_with_missing_rating_count('amazon.csv')

if missing_rating_data is not None:
  missing_rating_data
```

# 2.4 Removing missing data in rating_count column.

```python
# Remove missing_rating_data

def remove_missing_rating_data(file_path):
    try:
        df = pd.read_csv(file_path)
        # Remove rows where 'rating_count' is missing
        df_cleaned = df.dropna(subset=['rating_count'])
        return df_cleaned
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
        return None
    except KeyError:
        print(f"Error: 'rating_count' column not found in the file.")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

# Example usage
cleaned_data = remove_missing_rating_data('amazon.csv')

if cleaned_data is not None:
    cleaned_data.head()
```

## 2.6 Convert currency to dollars

```python
import pandas as pd

# Load the CSV file
file_path = 'amazon.csv'
data = pd.read_csv(file_path)

# Convert 'actual_price' and 'discounted_price' to numeric, handling potential errors
def to_numeric(x):
    try:
        # Attempt to convert to numeric, removing commas and currency symbols if present
        return pd.to_numeric(x.str.replace(',', '').str.replace('₹', ''))  # Assuming Indian Rupees
    except AttributeError:
        # If already numeric, return as is
        return x

data['actual_price'] = to_numeric(data['actual_price'])
data['discounted_price'] = to_numeric(data['discounted_price'])

# The currency conversion
currency_conversion_rate = 0.012  # Example: 1 INR = 0.012 USD
data['actual_price_usd'] = data['actual_price'] * currency_conversion_rate
data['discounted_price_usd'] = data['discounted_price'] * currency_conversion_rate

# Round 'actual_price_usd' and 'discounted_price_usd' to 2 decimal places
data['actual_price_usd'] = data['actual_price_usd'].round(2)
data['discounted_price_usd'] = data['discounted_price_usd'].round(2)

# Display the updated DataFrame with the new columns
print(data.head())
```

```python
# Sentiment analysis of words and low ratings

import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Assuming 'data' DataFrame is already loaded and preprocessed as in your previous code

# Download VADER lexicon if you haven't already
nltk.download('vader_lexicon')

# Initialize VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Function to get sentiment scores
def get_sentiment_scores(text):
    if isinstance(text, str):  # Check if the input is a string
        scores = analyzer.polarity_scores(text)
        return scores
    else:
        return None  # Return None for non-string values

# Apply sentiment analysis to reviews
data['sentiment_scores'] = data['review_content'].apply(get_sentiment_scores)

# Extract compound sentiment score
data['compound_sentiment'] = data['sentiment_scores'].apply(lambda x: x['compound'] if x else None)


# Analyze sentiment for low ratings
low_rating_threshold = 3  # Define the threshold for "low ratings"
low_rated_reviews = data[data['rating'] <= low_rating_threshold]

# Analyze the sentiment of low-rated reviews
low_rated_reviews['sentiment_scores'] = low_rated_reviews['review_content'].apply(get_sentiment_scores)

# Extract compound sentiment for low-rated reviews
low_rated_reviews['compound_sentiment'] = low_rated_reviews['sentiment_scores'].apply(lambda x: x['compound'] if x else None)


# Example: Print the reviews with low ratings and their sentiment scores
print(low_rated_reviews[['review_content', 'rating', 'compound_sentiment']])
```

## 9. Build Classification model that categorizes high and low rating outcomes.

```python
# Classification model to categorize outcomes high vs. low rating
# Inputs discounted_price, actual_price and rating_count
# Outputs rating_category of above 4 (hi) or below 4(low)

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier # Import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns


# Define a threshold for rating (e.g., 4.0)
rating_threshold = 4.0
data['rating_category'] = data['rating'].apply(lambda x: 'high' if x >= rating_threshold else 'low')

# Feature selection and target variable
features = data[['discounted_price', 'actual_price', 'rating_count']]
target = data['rating_category']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Initialize and train a RandomForestClassifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred))

# Create a confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['low', 'high'], yticklabels=['low', 'high'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

Conclusions
Strengths:
The model performs well for the "high" class, with high precision, recall, and F1-score. This indicates the model is reliable for detecting and predicting instances of this class.
Accuracy and weighted metrics suggest reasonable overall performance.
Weaknesses:
The model struggles with the "low" class, especially in recall (0.34). This indicates many true instances of the "low" class are missed.
Imbalance in class representation (support) likely contributes to the poor performance on the "low" class.
